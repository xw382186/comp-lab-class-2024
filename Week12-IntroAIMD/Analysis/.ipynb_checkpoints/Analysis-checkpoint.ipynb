```python
import os
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
from scipy.ndimage import gaussian_filter1d

# 设置路径
run_path = "/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week12-IntroAIMD/Runs/Proton"
figure_path = "/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week12-IntroAIMD/Figure"

# 1. 读取能量 vs 时间
ener_file = os.path.join(run_path, "proton_aimd_1ns_every0.5fs-1.ener")
energy_data = []

with open(ener_file, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if not line.startswith("#"):
            parts = line.split()
            time = float(parts[0])  # 时间
            energy = float(parts[1])  # 能量
            energy_data.append((time, energy))

time, energy = zip(*energy_data)

plt.figure(figsize=(10, 6))
plt.plot(time, energy, label="Energy vs Time", color="blue", linestyle="--", linewidth=1.5)
plt.xlabel("Time (fs)", fontsize=16)
plt.ylabel("Energy (Hartree)", fontsize=16)
plt.title("Energy vs Time", fontsize=18)
plt.legend(fontsize=14)
plt.grid(alpha=0.3)
plt.savefig(os.path.join(figure_path, "energy_vs_time.png"), dpi=300)
plt.show()

# 2. 计算氧原子的 g(r)
xyz_file = os.path.join(run_path, "proton_aimd_1ns_every0.5fs-pos-1.xyz")
box_size = [12.42, 12.42, 12.42]  # 从输入文件获得盒子大小

# 手动加载拓扑信息
with open(xyz_file, 'r') as f:
    lines = f.readlines()
    num_atoms = int(lines[0])
    atom_names = [line.split()[0] for line in lines[2:2 + num_atoms]]
    positions = np.array([[float(x) for x in line.split()[1:4]] for line in lines[2:2 + num_atoms]])

# 创建 Topology 和 Trajectory 对象
topology = md.Topology()
chain = topology.add_chain()
residue = topology.add_residue("MOL", chain)
for atom_name in atom_names:
    topology.add_atom(atom_name, element=None, residue=residue)

traj = md.Trajectory([positions], topology=topology)

# 为 traj 设置 3x3 单位盒子向量
unitcell_vectors = np.array([
    [[box_size[0], 0, 0],
     [0, box_size[1], 0],
     [0, 0, box_size[2]]] for _ in range(traj.n_frames)
])
traj.unitcell_vectors = unitcell_vectors

# 计算 RDF
oxygen_indices = traj.top.select('name O')  # 选择氧原子
pairs = np.array([[i, j] for i in oxygen_indices for j in oxygen_indices if i < j])

rdf, r = md.compute_rdf(traj, pairs, r_range=(2.0, 8.0), bin_width=0.01)
rdf_smooth = gaussian_filter1d(rdf, sigma=3)  # 增强平滑效果

# 绘制 RDF 图
plt.figure(figsize=(10, 8))
plt.plot(r, rdf_smooth, label="Simulated RDF (300K)", linestyle="-", color="darkblue", linewidth=2)
plt.xlabel(r"$r_{OO}$ ($\mathrm{\AA}$)", fontsize=16)
plt.ylabel(r"$g(r_{OO})$", fontsize=16)
plt.title("Radial Distribution Function (RDF) for Oxygen", fontsize=18)
plt.xlim(2, 8)
plt.ylim(0, 8)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.grid(alpha=0.5, linestyle="--")
plt.savefig(os.path.join(figure_path, "oxygen_rdf_reference_style.png"), dpi=300)
plt.show()
```
