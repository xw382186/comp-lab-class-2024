```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import PandasTools
from psikit import Psikit

# 读取 SMILES 数据并将其转换为 DataFrame
smiles_df = pd.read_csv("smiles.csv")

# 添加 ROMol 列，将 SMILES 转换为 RDKit 分子对象
PandasTools.AddMoleculeColumnToFrame(smiles_df, smilesCol='smiles')

# 添加 n_atoms 列，记录每个分子的原子数
smiles_df['n_atoms'] = smiles_df['ROMol'].map(lambda x: x.GetNumAtoms() if x else None)
smiles_df.head()
```

```python
# 初始化 Psikit
pk = Psikit(debug=True, threads=1, memory=12)

# 计算每个分子的 SCF 能量，并记录在 Energy 列
energies = []
for smiles in smiles_df['smiles']:
    pk.read_from_smiles(smiles)
    energy = pk.optimize()
    energies.append(energy)

# 添加 Energy 列到 DataFrame
smiles_df['Energy'] = energies
smiles_df.head()
```
