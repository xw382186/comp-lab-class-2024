```python
import matplotlib.pyplot as plt

# Define a function to parse .xvg files, skipping comments and extracting data
def parse_xvg_custom(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            # Skip lines that are comments (starting with # or @)
            if not line.startswith(('#', '@')):
                # Try splitting by spaces or tabs
                cols = line.split()
                try:
                    # Ensure we have five columns and convert to floats
                    if len(cols) == 5:
                        data.append([float(cols[0]), float(cols[1])])
                except ValueError:
                    # Skip lines that cannot be converted to floats
                    continue
    return data

# File path to the gyration data
gyrate_file_path = './gyrate.xvg'

# Parse gyration data (time and total gyration radius)
data_gyrate = parse_xvg_custom(gyrate_file_path)

# Extract time and gyration radius for plotting
time_gyrate, gyration_total = zip(*data_gyrate)

# Plotting the total gyration radius over time
plt.figure(figsize=(10, 6))
plt.plot(time_gyrate, gyration_total, label="Total Gyration Radius", color='green')
plt.title("Total Gyration Radius Over Time")
plt.xlabel("Time (ns)")  # Set x-axis label with (ns)
plt.ylabel("Gyration Radius (nm)")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
