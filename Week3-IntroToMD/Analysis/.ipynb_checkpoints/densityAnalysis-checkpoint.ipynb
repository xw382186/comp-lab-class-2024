```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path to your .xvg file
file_path = './density.xvg'  # Adjust path if necessary

# Load the data
with open(file_path) as file:
    lines = file.readlines()

# Parse the .xvg file, excluding comment lines
data = []
for line in lines:
    if not line.startswith(('#', '@')):
        data.append([float(x) for x in line.split()])

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(data, columns=['Time (ps)', 'Density (kg/m^3)'])

# Plot the density over time
plt.figure(figsize=(10,6))
plt.plot(df['Time (ps)'], df['Density (kg/m^3)'], color='green', label='Density')
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m^3)')
plt.title('Density vs Time in GROMACS Simulation')
plt.grid(True)
plt.legend()
plt.show()

# Save the plot as an image (optional)
plt.savefig("density_analysis_plot.png")
```
