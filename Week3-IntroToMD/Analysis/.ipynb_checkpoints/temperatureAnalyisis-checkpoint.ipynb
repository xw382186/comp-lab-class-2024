```python
# 
import matplotlib.pyplot as plt
import numpy as np

# 
def read_temperature_data(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            if not line.startswith(('@', '#')):  # 
                data.append(list(map(float, line.split())))
    return data

# 
def plot_temperature(time, temperature):
    plt.figure(figsize=(10, 6))
    plt.plot(time, temperature, label='Temperature (K)', color='r')
    plt.xlabel('Time (ps)')
    plt.ylabel('Temperature (K)')
    plt.title('Temperature Equilibration Over Time')
    plt.grid(True)
    plt.legend()
    plt.show()

# 
def calculate_temperature_stats(temperature):
    mean_temp = np.mean(temperature)
    max_temp = np.max(temperature)
    min_temp = np.min(temperature)
    return mean_temp, max_temp, min_temp

# 
file_path = './temperature.xvg'  # temperature.xvg

# 
temperature_data = read_temperature_data(file_path)

# 
time = [row[0] for row in temperature_data]  # 1 (ps)
temperature = [row[1] for row in temperature_data]  # 2 (K)

# 
plot_temperature(time, temperature)

# 
mean_temp, max_temp, min_temp = calculate_temperature_stats(temperature)
print(f": {mean_temp:.2f} K")
print(f": {max_temp:.2f} K")
print(f": {min_temp:.2f} K")
```
