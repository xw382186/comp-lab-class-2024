```python
import matplotlib.pyplot as plt

# Define a function to parse .xvg files, skipping comments and extracting data
def parse_xvg(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            # Skip comment lines (starting with # or @)
            if not line.startswith(('#', '@')):
                cols = line.split()  # Split each line by spaces
                if len(cols) == 2:  # Ensure each line has two data points
                    data.append([float(cols[0]), float(cols[1])])
    return data

# File paths (adjust to your file locations)
file_xtal_path = './rmsd_xtal.xvg'
file_rmsd_path = './rmsd.xvg'

# Parse data from both files
data_xtal = parse_xvg(file_xtal_path)
data_rmsd = parse_xvg(file_rmsd_path)

# Convert data to lists for plotting
time_xtal, rmsd_xtal = zip(*data_xtal)
time_rmsd, rmsd = zip(*data_rmsd)

# Plot RMSD data
plt.figure(figsize=(10, 6))
plt.plot(time_xtal, rmsd_xtal, label="RMSD_xtal", color='blue')
plt.plot(time_rmsd, rmsd, label="RMSD", color='orange')
plt.title("RMSD Comparison Over Time")
plt.xlabel("Time (ns)")  # Set x-axis label with (ns)
plt.ylabel("RMSD (nm)")
plt.legend()
plt.grid(True)

# Display plot
plt.show()


```
