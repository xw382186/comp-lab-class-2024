```python
# 导入必要的库
import matplotlib.pyplot as plt

# 读取数据文件，过滤掉注释行
file_path = './potential.xvg'  # 请将此路径替换为实际文件路径

data = []
with open(file_path, 'r') as f:
    for line in f:
        if not line.startswith(('@', '#')):
            data.append(list(map(float, line.split())))

# 提取步数和势能列
steps = [row[0] for row in data]  # 步数列
potential_energy = [row[1] for row in data]  # 势能列

# 绘制能量最小化曲线
plt.figure(figsize=(10, 6))
plt.plot(steps, potential_energy, label='Potential Energy', color='b')
plt.xlabel('Steps', fontsize=12)
plt.ylabel('Potential Energy (kJ/mol)', fontsize=12)
plt.title('Energy Minimization Curve', fontsize=14)
plt.grid(True)
plt.legend()

# 显示图表
plt.show()

# 打印最后的能量值作为收敛判断
print(f"最后一步的势能: {potential_energy[-1]} kJ/mol")

# 能量差异分析
initial_energy = potential_energy[0]
final_energy = potential_energy[-1]
energy_difference = initial_energy - final_energy

print(f"初始势能: {initial_energy} kJ/mol")
print(f"最终势能: {final_energy} kJ/mol")
print(f"能量变化（最小化过程）: {energy_difference} kJ/mol")
```
