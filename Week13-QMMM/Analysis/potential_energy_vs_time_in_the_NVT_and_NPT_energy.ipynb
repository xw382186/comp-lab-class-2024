```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import os

# Define file paths
npt_file = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/NPT-1.ener'
nvt_file = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/NVT-1.ener'

# Load NPT data
npt_data = np.loadtxt(npt_file, skiprows=1)
time_npt_fs = npt_data[:, 1]  # Time in fs
potential_energy_npt_au = npt_data[:, 4]  # Potential Energy in a.u.

# Load NVT data
nvt_data = np.loadtxt(nvt_file, skiprows=1)
time_nvt_fs = nvt_data[:, 1]  # Time in fs
potential_energy_nvt_au = nvt_data[:, 4]  # Potential Energy in a.u.

# Convert units
time_npt_ps = time_npt_fs / 1000  # Convert fs to ps
potential_energy_npt_kcal = potential_energy_npt_au * 627.509  # Convert a.u. to kcal/mol

time_nvt_ps = time_nvt_fs / 1000  # Convert fs to ps
potential_energy_nvt_kcal = potential_energy_nvt_au * 627.509  # Convert a.u. to kcal/mol

# Plot potential energy vs time
plt.figure(figsize=(10, 6))

# Plot NPT data
plt.plot(time_npt_ps, potential_energy_npt_kcal, label='NPT Potential Energy', color='blue')

# Plot NVT data
plt.plot(time_nvt_ps, potential_energy_nvt_kcal, label='NVT Potential Energy', color='green')

# Add labels, title, and legend
plt.xlabel('Time (ps)', fontsize=14)
plt.ylabel('Potential Energy (kcal/mol)', fontsize=14)
plt.title('Potential Energy vs Time (NPT & NVT)', fontsize=16)
plt.legend(fontsize=12)

# Ensure output directory exists
output_dir = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Figures'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'npt_nvt_potential_energy_vs_time.png')

# Save the plot
plt.savefig(output_path, dpi=300)
plt.show()

# Print confirmation
print(f"Plot saved to {output_path}")
```
