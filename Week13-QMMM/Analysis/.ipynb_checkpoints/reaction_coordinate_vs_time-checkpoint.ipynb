```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 文件路径
cv_file = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/monitor/MONITOR-COLVAR.metadynLog'

# 读取数据
cv_data = np.loadtxt(cv_file)
time_fs = cv_data[:, 0]  # 时间（飞秒）
reaction_coordinate = cv_data[:, 1]  # 反应坐标

# 单位转换
time_ps = time_fs / 1000  # 飞秒转皮秒

# 绘制反应坐标 vs 时间
plt.figure(figsize=(10, 6))
plt.plot(time_ps, reaction_coordinate, label='Reaction Coordinate', color='purple')

# 添加标签和图例
plt.xlabel('Time (ps)', fontsize=14)
plt.ylabel('Reaction Coordinate (Å)', fontsize=14)  # 假设反应坐标是距离
plt.title('Reaction Coordinate vs Time', fontsize=16)
plt.legend(fontsize=12)

# 确保输出目录存在
output_dir = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Figures'
os.makedirs(output_dir, exist_ok=True)

# 保存图片
output_path = os.path.join(output_dir, 'reaction_coordinate_vs_time.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Reaction Coordinate plot saved at: {output_path}")
```
