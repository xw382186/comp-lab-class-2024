```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 
cv_files = [
    '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog',
    '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog'
]

# 
labels = ['Metadynamics Standard', 'Metadynamics Modified']
colors = ['blue', 'green']

#  vs 
plt.figure(figsize=(10, 6))

for i, cv_file in enumerate(cv_files):
    # 
    data = np.loadtxt(cv_file)
    time_fs = data[:, 0]  # 
    reaction_coordinate = data[:, 1]  # 

    # 
    time_ps = time_fs / 1000  # 

    # 
    plt.plot(time_ps, reaction_coordinate, label=labels[i], color=colors[i])

# 
plt.xlabel('Time (ps)', fontsize=14)
plt.ylabel('Reaction Coordinate', fontsize=14)
plt.title('Reaction Coordinate vs Time (Metadynamics)', fontsize=16)
plt.legend(fontsize=12)

# 
output_dir = '/scratch/wx2237/comp-lab-class/comp-lab-class-2024/Week13-QMMM/Figures'
os.makedirs(output_dir, exist_ok=True)

# 
output_path = os.path.join(output_dir, 'reaction_coordinate_vs_time_metadynamics.png')
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Reaction Coordinate plot saved at: {output_path}")
```
