```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt

# 加载轨迹文件，确保使用新的 tpr 和 xtc 文件
u = mda.Universe('protein_only.tpr', 'protein_only.xtc')

# 选择蛋白质的第一个和最后一个原子
start_atom = u.select_atoms("bynum 1")  # 第一个原子
end_atom = u.select_atoms("bynum 501")  # 最后一个原子，假设是第501个

# 创建列表来存储时间和距离
times = []
distances = []

# 遍历轨迹，计算每一帧的末端距离
for ts in u.trajectory:
    distance = np.linalg.norm(start_atom.positions - end_atom.positions)
    times.append(ts.time)
    distances.append(distance)

# 将距离数据绘制出来
plt.plot(times, distances, label='End-to-End Distance')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (Å)')
plt.title('End-to-End Distance of Protein Over Time')
plt.grid(True)
plt.legend()
plt.show()
```
