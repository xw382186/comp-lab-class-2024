```python
import matplotlib.pyplot as plt
import numpy as np

#  rmsd 
rmsd_data = np.loadtxt('rmsd.xvg', comments=['@', '#'])

#  (ps)  RMSD (nm)
time = rmsd_data[:, 0]
rmsd = rmsd_data[:, 1]

# 
plt.figure(figsize=(8, 6))
plt.plot(time, rmsd, label='RMSD', color='b')

# 
plt.title('RMSD vs Time (ps)')
plt.xlabel('Time (ps)')
plt.ylabel('RMSD (nm)')
plt.legend()

#  PNG 
plt.savefig('RMSD_plot.png')

# 
plt.show()
```
