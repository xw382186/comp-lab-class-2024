```python
import matplotlib.pyplot as plt
import numpy as np

# 读取 rmsd 数据文件
rmsd_data = np.loadtxt('rmsd.xvg', comments=['@', '#'])

# 提取时间 (ps) 和 RMSD (nm)
time = rmsd_data[:, 0]
rmsd = rmsd_data[:, 1]

# 创建绘图
plt.figure(figsize=(8, 6))
plt.plot(time, rmsd, label='RMSD', color='b')

# 添加标题和标签
plt.title('RMSD vs Time (ps)')
plt.xlabel('Time (ps)')
plt.ylabel('RMSD (nm)')
plt.legend()

# 保存图像为 PNG 文件
plt.savefig('RMSD_plot.png')

# 显示图形
plt.show()
```
