```python
import MDAnalysis as mda
from MDAnalysis.analysis.msd import EinsteinMSD
import matplotlib.pyplot as plt

# 加载轨迹文件和拓扑文件
u = mda.Universe('step5_5.tpr', 'step5_5.xtc')

# 选择水的氧原子 (OH2)，钠离子 (SOD) 和氯离子 (CLA)
water_oxygens = u.select_atoms("name OH2")
sodium = u.select_atoms("name SOD")
chlorine = u.select_atoms("name CLA")

# 打印确认原子数量
print(f"Number of water oxygens: {len(water_oxygens)}")
print(f"Number of sodium atoms: {len(sodium)}")
print(f"Number of chlorine atoms: {len(chlorine)}")

# 计算 MSD (均方位移)
msd_water = EinsteinMSD(water_oxygens, select='all', msd_type='xyz').run()
msd_sodium = EinsteinMSD(sodium, select='all', msd_type='xyz').run()
msd_chlorine = EinsteinMSD(chlorine, select='all', msd_type='xyz').run()

# 绘制 MSD 与时间的关系图
plt.plot(msd_water.times, msd_water.results.timeseries, label='Water Oxygen')
plt.plot(msd_sodium.times, msd_sodium.results.timeseries, label='Sodium')
plt.plot(msd_chlorine.times, msd_chlorine.results.timeseries, label='Chlorine')

plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm^2)')
plt.legend()
plt.title('MSD vs Time for Water Oxygens, Sodium, and Chlorine')
plt.savefig('Figures/MSD_plot.png')
plt.show()

```
