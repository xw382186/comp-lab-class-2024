```python
# 导入所需的库
import numpy as np
import matplotlib.pyplot as plt

# 定义函数以加载能量数据，跳过注释行
def load_energy_data(file_path):
    data = np.loadtxt(file_path, comments=['@', '#'])  # 跳过以 @ 和 # 开头的行
    return data[:, 1]  # 返回势能数据

# 1. 计算势能直方图
# 加载各温度下的势能数据
energy_t300 = load_energy_data('T300/potential_energy.xvg')
energy_t400 = load_energy_data('T400/potential_energy.xvg')
energy_t600 = load_energy_data('T600/potential_energy.xvg')

# 绘制能量直方图
plt.figure(figsize=(10, 6))
plt.hist(energy_t300, bins=50, alpha=0.5, label='T=300 K', density=True)
plt.hist(energy_t400, bins=50, alpha=0.5, label='T=400 K', density=True)
plt.hist(energy_t600, bins=50, alpha=0.5, label='T=600 K', density=True)

plt.title('Energy Histograms at Different Temperatures')
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.savefig('energy_histograms.png')  # 保存直方图
plt.show()

```
