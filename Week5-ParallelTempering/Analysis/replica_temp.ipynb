```python
import numpy as np
import matplotlib.pyplot as plt

# 
def load_temperature_data(file_path):
    data = np.loadtxt(file_path, comments=['@', '#'])  #  @  # 
    return data[:, 0], data[:, 1]  # 

#  T=300 
time_t300, temp_t300 = load_temperature_data('T300/replica_temp.xvg')

#  T=400 
time_t400, temp_t400 = load_temperature_data('T400/replica_temp.xvg')

#  T=600 
time_t600, temp_t600 = load_temperature_data('T600/replica_temp.xvg')

# 100
sampling_interval = 100
time_sampled_t300 = time_t300[::sampling_interval]
temp_sampled_t300 = temp_t300[::sampling_interval]

time_sampled_t400 = time_t400[::sampling_interval]
temp_sampled_t400 = temp_t400[::sampling_interval]

time_sampled_t600 = time_t600[::sampling_interval]
temp_sampled_t600 = temp_t600[::sampling_interval]

# 
plt.figure(figsize=(12, 8))

#  T=300
plt.subplot(3, 1, 1)
plt.plot(time_sampled_t300, temp_sampled_t300, label='Replica Temperatures', color='b')
plt.title('Temperature Variation of Replicas at T=300 K')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()

#  T=400
plt.subplot(3, 1, 2)
plt.plot(time_sampled_t400, temp_sampled_t400, label='Replica Temperatures', color='g')
plt.title('Temperature Variation of Replicas at T=400 K')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()

#  T=600
plt.subplot(3, 1, 3)
plt.plot(time_sampled_t600, temp_sampled_t600, label='Replica Temperatures', color='r')
plt.title('Temperature Variation of Replicas at T=600 K')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.savefig('replica_temperature_variation_all.png')  # 
plt.show()





```
