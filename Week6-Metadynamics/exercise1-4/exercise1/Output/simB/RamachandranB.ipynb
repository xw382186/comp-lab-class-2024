```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Load COLVAR data
# You can adjust the file path accordingly if needed
data = np.loadtxt('COLVAR')

# Extract time, phi, and psi columns
time = data[:, 0]
phi = data[:, 1]
psi = data[:, 2]

# Plot Phi and Psi over time
plt.figure(figsize=(10, 5))
plt.plot(time, phi, label='Phi (radians)', color='blue')
plt.plot(time, psi, label='Psi (radians)', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (radians)')
plt.title('Phi and Psi Dihedral Angles over Time')
plt.legend()
plt.grid(True)
plt.show()

# Compute standard deviations of Phi and Psi
phi_std = np.std(phi)
psi_std = np.std(psi)

print(f'Standard deviation of Phi: {phi_std:.4f} radians')
print(f'Standard deviation of Psi: {psi_std:.4f} radians')

# Ramachandran Plot (phi vs psi)
plt.figure(figsize=(6, 6))
plt.scatter(phi, psi, s=10, color='purple')
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('Ramachandran Plot (Phi vs Psi)')
plt.grid(True)
plt.show()

# Optional: Convert radians to degrees for further analysis if needed
phi_deg = np.degrees(phi)
psi_deg = np.degrees(psi)

# Ramachandran Plot in degrees
plt.figure(figsize=(6, 6))
plt.scatter(phi_deg, psi_deg, s=10, color='red')
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.title('Ramachandran Plot (Phi vs Psi) in Degrees')
plt.grid(True)
plt.show()
```
