```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

# 加载 AF2 (CIF 文件) 和 AF3 (PDB 文件) 的结构
af2 = md.load('AF2-actin.pdb', top='AF2-actin.pdb')
af3 = md.load('AF3-actin.pdb', top='AF3-actin.pdb')

# 选择链A中的所有原子
af2_chainA = af2.atom_slice(af2.topology.select('chainid 0'))
af3_chainA = af3.atom_slice(af3.topology.select('chainid 0'))

# 对齐 AF2 和 AF3 链A
af3_chainA.superpose(af2_chainA)

# 计算每个残基的均方根距离（RMSD）
residue_distances = []
for res in af2_chainA.topology.residues:
    af2_residue_atoms = af2_chainA.topology.select(f'resid {res.index}')
    af3_residue_atoms = af3_chainA.topology.select(f'resid {res.index}')
    
    # 计算每个残基的质心
    af2_residue_center = af2_chainA.xyz[:, af2_residue_atoms].mean(axis=1)
    af3_residue_center = af3_chainA.xyz[:, af3_residue_atoms].mean(axis=1)
    
    # 计算两个模型中该残基质心之间的距离
    distance = np.linalg.norm(af2_residue_center - af3_residue_center)
    residue_distances.append(distance)

# 绘制距离 vs 残基号的条形图
residue_numbers = [res.index for res in af2_chainA.topology.residues]
plt.bar(residue_numbers, residue_distances)
plt.xlabel('Residue Number')
plt.ylabel('Distance (nm)')
plt.title('Distance between AF2 and AF3 predictions (Actin Chain A)')
plt.show()

# 标记距离较大的区域
threshold = 0.2  # 可根据需要调整
different_regions = [i for i, dist in enumerate(residue_distances) if dist > threshold]
print(f"Regions with significant differences: {different_regions}")
```
