```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = './1hz3_T310.run.25000000.energy.xvg'

#  @  # 
data_lines = []
with open(file_path, 'r') as file:
    for line in file:
        if not line.startswith(('@', '#')):  # 
            data_lines.append(line.split())

# DataFrame7
data = pd.DataFrame([line for line in data_lines if len(line) == 7])

# 
data = data.apply(pd.to_numeric)

# 
data.columns = ['Frame', 'Potential', 'Kinetic En.', 'Total Energy', 'Temperature', 'Pressure', 'Volume']

# 050000
frames_0_to_50000 = data[data['Frame'] <= 50000]

#  Potential 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Potential'], label='Potential')
plt.xlabel('Frame')
plt.ylabel('Potential Energy')
plt.title('Potential Energy vs Frame (0-50000)')
plt.legend()
plt.show()

#  Kinetic En. 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Kinetic En.'], label='Kinetic En.')
plt.xlabel('Frame')
plt.ylabel('Kinetic Energy')
plt.title('Kinetic Energy vs Frame (0-50000)')
plt.legend()
plt.show()

#  Total Energy 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Total Energy'], label='Total Energy')
plt.xlabel('Frame')
plt.ylabel('Total Energy')
plt.title('Total Energy vs Frame (0-50000)')
plt.legend()
plt.show()

#  Temperature 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Temperature'], label='Temperature')
plt.xlabel('Frame')
plt.ylabel('Temperature (K)')
plt.title('Temperature vs Frame (0-50000)')
plt.legend()
plt.show()

#  Pressure 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Pressure'], label='Pressure')
plt.xlabel('Frame')
plt.ylabel('Pressure (bar)')
plt.title('Pressure vs Frame (0-50000)')
plt.legend()
plt.show()

#  Volume 
plt.figure(figsize=(10, 6))
plt.plot(frames_0_to_50000['Frame'], frames_0_to_50000['Volume'], label='Volume')
plt.xlabel('Frame')
plt.ylabel('Volume (nm^3)')
plt.title('Volume vs Frame (0-50000)')
plt.legend()
plt.show()
```
